# clustering- uni, bi, multivariate

```python
clustering1 = KMeans(n_clusters=6)
```

```python
clustering1.fit(customer_data[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
customer_data['Income Cluster'] = clustering1.labels_
customer_data.head()
```

```python
customer_data['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
inertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(customer_data[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
```

```python
inertia_scores
```

```python
plt.plot(range(1,11),inertia_scores)
```
